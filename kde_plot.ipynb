```python
import pandas as pd
import seaborn as sns

# Assuming 'crash_df' contains your data
# Create a new categorical variable indicating alcohol level ranges
crash_df['alcohol_range'] = pd.cut(crash_df['alcohol'], bins=[0, 0.05, 0.1, 0.15, float('inf')], labels=['None', 'Low', 'Medium', 'High'])

# Plotting with hue as the new 'alcohol_range' variable
sns.kdeplot(data=crash_df, x='speeding', hue='alcohol_range', fill=True)
```

In your context, the bivariate KDE plot with x='speeding' and y='not_distracted', colored by hue='alcohol', would help in understanding how the speeding and distraction levels are distributed and how they differ based on alcohol involvement.

A KDE plot represents the data using a continuous probability density curve. It is useful for visualizing the shape of the data distribution without being affected by the bin size, as in a histogram.

# Key Components of KDE Plot


Kernel Function: Determines the shape of the smooth curve used to create the density estimate. Common kernels include Gaussian, Epanechnikov, and Tophat.
Bandwidth: Controls the smoothness of the KDE curve. A smaller bandwidth leads to a more sensitive estimate with more peaks and valleys, while a larger bandwidth produces a smoother estimate.

# When to Use KDE Plot
Understanding Distribution: When you want to understand the distribution of your data, especially if you suspect it may not be normally distributed.
Comparing Distributions: Useful for comparing the distributions of multiple datasets.
Visualizing Data Smoothly: When you need a smooth representation of the data that avoids the potential noise introduced by histogram binning

# Key Parameters in kdeplot
data: DataFrame containing the data.

x, y: Variables to plot.

bw_adjust: Factor to adjust the bandwidth.

fill: Whether to fill the area under the KDE curve.

cumulative: Whether to compute a cumulative distribution.

hue: Semantic variable to create multiple KDE plots.

palette: Color palette to use for the different levels of the hue variable.

shade: (For bivariate KDE) Whether to shade the KDE plot.
