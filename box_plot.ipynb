```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
data = sns.load_dataset("tips")

# Create a basic box plot
sns.boxplot(x="day", y="tip", data=df)

# Show plot
plt.title("Box Plot of Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.show()
```

```python
# Box plot with hue
sns.boxplot(x="day", y="total_bill", hue="time", data=df)

# Show plot
plt.title("Box Plot of Total Bill by Day and Time")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.legend(title="Time of Day")
plt.show()
```

```python
# Horizontal box plot
sns.boxplot(y="day", x="total_bill", data=data)

# Show plot
plt.title("Box Plot of Total Bill by Day")
plt.xlabel("Total Bill ($)")
plt.ylabel("Day of the Week")
plt.show()
```

A box plot, also known as a box-and-whisker plot, is a graphical summary of the distribution of a numerical variable through its quartiles. It provides a concise way to visualize the central tendency, variability, and skewness of the data. Box plots are particularly useful for identifying outliers and comparing distributions across different categories or groups.

# Features of Box Plot
Five-Number Summary: The box plot summarizes the distribution using five statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum.

Interquartile Range (IQR): The range between the first and third quartiles, representing the middle 50% of the data.

Whiskers: Extend from the box to the minimum and maximum values within a certain distance from the quartiles (typically 1.5 times the IQR).

Outliers: Data points beyond the whiskers are considered outliers and are plotted individually.

# Key Parameters
x and y: Variables to be plotted on the x-axis and y-axis, respectively.
data: DataFrame containing the data to be plotted.

hue: Variable for grouping the data and plotting separate box plots for each group.

order and hue_order: Order to plot the categories along the x-axis or hue variable.

palette: Color palette for different categories or groups.

notch: Whether to draw a notch around the median to provide a confidence interval for the median.

# Use Cases for Box Plots
Comparing Distributions: Box plots are effective for comparing the distributions of numerical variables across different categories or groups.

Identifying Outliers: They help in identifying outliers or extreme values in the data.

Detecting Skewness: Box plots can reveal the skewness of the data distribution based on the position and length of the whiskers.

Visualizing Quartiles: They provide a visual representation of the quartiles and central tendency of the data.

Assessing Variability: Box plots aid in understanding the variability and spread of the data.

# When to Use Box Plots
Comparing Groups: When you need to compare the distribution of a numerical variable across different categories or groups.

Identifying Outliers: When you want to identify outliers or extreme values in the data.

Understanding Variability: When you need to understand the variability and spread of the data.

Communicating Summary Statistics: When summarizing the distribution using quartiles and central tendency is important.
