```python
import numpy as pd
import pandas as pd 
import matplotlib as plt
import seaborn as sns
# inline is used to plot within this notebook
%matplotlib inline
```

```python
#to get inbuild data_sets from the seaborn itself then :
sns.get_dataset_names()
```

```python
crash_df = sns.load_dataset('car_crashes')
```

```python
crash_df
```

```python

sns.distplot(crash_df['not_distracted'], kde = False)  
```

```python
bins = [5,15,20,30,40]
sns.distplot(crash_df['not_distracted'], bins = bins, hist_kws={'edgecolor':'k'})  
```

```python
bins = [5,15,20,30,40]
sns.distplot(crash_df['not_distracted'], bins = bins,hist_kws={'edgecolor':'k'})  
```

 # Histogram: Shows the frequency of data points within certain ranges (bins).
# KDE: A smoothed estimate of the distribution, useful for understanding the density of data points.
# hist_kws: A dictionary of keyword arguments to customize the histogram (e.g., edgecolor for bin edges).

# When to Use distplot
# Visualizing Distribution: Use distplot when you want a quick visualization of the distribution of a dataset, including both its histogram and KDE.
# Small to Medium Data: Suitable for datasets where combining histogram and KDE provides meaningful insights.
# Exploratory Data Analysis: Great for initial exploration of data to understand its distribution and density.
