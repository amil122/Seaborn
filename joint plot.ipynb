# jointplot is another powerful plotting function from seaborn that allows you to explore the relationship between two variables by displaying a bivariate plot along with univariate marginal distributions


# jointplot creates a multi-panel plot that shows:

# 1. A scatter plot (or other bivariate plot) of two variables.
# 2. Histograms or KDE plots of each variable on the axes.

# Bivariate Plot: This is the main plot in the center showing the relationship between the two variables.
# Default is a scatter plot.
# Can also be hexbin, KDE, regression, or residual plot.
# Marginal Distributions: These are the histograms or KDE plots on the top and right sides of the main plot, showing the univariate distribution of each variable.


# Parameters of jointplot


#x, y: Names of variables in the DataFrame for the x and y axes.
#data: DataFrame containing the data.
#kind: Type of plot to draw ("scatter", "reg", "resid", "kde", or "hex"). #Default is "scatter".
#height: Size of the figure (it will be square). Default is 6.
#ratio: Ratio of the size of the marginal axes to the main axes. Default is 5.
#space: Space between the main plot and the marginal plots. Default is 0.2.
#marginal_kws: Dictionary of keyword arguments to customize the marginal plots.
#joint_kws: Dictionary of keyword arguments to customize the bivariate plot.

# When to Use jointplot


Exploring Relationships: Use jointplot when you want to explore the relationship between two continuous variables, along with their individual distributions.
Visualizing Correlations: Helpful for identifying patterns, correlations, and_potential outliers.
Bivariate Analysis: Ideal for bivariate analysis in exploratory data analysis (EDA).

marginal_kws
the marginal_kws parameter in jointplot allows you to customize the marginal plots, which are the univariate distributions (histograms or KDE plots) shown on the top and right side of the main bivariate plot.
