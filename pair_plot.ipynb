Diagonal Plots: Along the diagonal of the pair plot matrix, histograms or kernel density estimates (KDEs) are typically displayed for each variable. These plots show the univariate distribution of each variable.

Off-diagonal Plots: Off-diagonal elements display scatterplots of paired variables, enabling visual inspection of their relationship.

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
iris = sns.load_dataset("iris")

# Basic pair plot with histograms on the diagonal
sns.pairplot(data=iris, diag_kind='hist')
plt.show()

# Pair plot with KDEs on the diagonal
sns.pairplot(data=iris, diag_kind='kde')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
penguins = sns.load_dataset("penguins")

# Drop rows with missing values to simplify the example
penguins = penguins.dropna()

# Select only the numerical columns for the pair plot matrix
numerical_cols = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']

# Create a pair plot using 'species' as the hue
sns.pairplot(data=penguins, vars=numerical_cols, hue='species')
plt.show()
```

In seaborn, the hue parameter in pairplot allows you to color the data points based on a categorical variable, but it doesn't directly support multiple hues. However, you can use a combination of hue and additional plot customization to visually distinguish between different categories.

Here is an example demonstrating how to use a single hue parameter. Unfortunately, seaborn does not natively support multiple hues in the pairplot function. If you need to handle more complex cases, you might need to use FacetGrid or create custom plots for each pair.

 the context of the Titanic dataset, if you want to compare the age and pclass while using a combined version of age and pclass for the hue, you would need to create a new categorical variable that represents the combination of age and pclass. However, typically age is a continuous variable, so combining it with pclass (a categorical variable) to create a hue variable might not make practical sense, since hue is usually a categorical variable.

Instead, you could use pclass as the hue directly and compare age with other numerical variables. If you still want to combine age and pclass for some specific reason, you can create a new variable that bins age into categories and then combine these bins with pclass.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Drop rows with missing values in 'age' and 'pclass' to simplify the example
titanic = titanic.dropna(subset=['age', 'pclass'])

# Select numerical columns for the pair plot
numerical_cols = ['age', 'fare']

# Create a pair plot using 'pclass' as the hue
sns.pairplot(data=titanic, vars=numerical_cols, hue='pclass', diag_kind='kde')
plt.show()
```
