# Plots for Numerical Variables

# Histplot

Description: Displays the distribution of a single numerical variable as a histogram.

Use Case: Understanding the frequency distribution of a variable.

Example:
sns.histplot(data=df, x='numerical_column')

# KDE Plot (Kernel Density Estimate):
Description: Estimates the probability density function of a numerical variable.

Use Case: Visualizing the distribution of a variable smoothly.
sns.kdeplot(data=df, x='numerical_column')

# Boxplot
Description: Displays the distribution of a numerical variable through its quartiles.

Use Case: Identifying outliers and the spread of data.
    sns.boxplot(y='numerical_column', data=df)

# Violin Plot
Description: Combines a boxplot with a KDE plot to show the distribution and probability density of a numerical variable.

Use Case: Understanding the distribution and density of data.
Example:
sns.violinplot(y='numerical_column', data=df)

# Plots for Numerical-Numerical Variables

# Scatterplot
Description: Plots individual data points based on two numerical variables.

Use Case: Identifying relationships and correlations between two numerical variables.
Example:
sns.scatterplot(x='numerical_column1', y='numerical_column2', data=df)

# Lineplot
Description: Connects data points with lines based on two numerical variables.

Use Case: Showing trends over time or ordered numerical data.
Example:
sns.lineplot(x='numerical_column1', y='numerical_column2', data=df)

# Jointplot
Description: Combines scatterplot and histograms (or KDE plots) to show the relationship between two numerical variables.

Use Case: Analyzing the relationship and distribution of two numerical variables.
Example:
sns.jointplot(x='numerical_column1', y='numerical_column2', data=df)

# Pairplot
Description: Creates a grid of scatterplots for all pairs of numerical variables.

Use Case: Exploring relationships and distributions among multiple numerical variables.
Example:
sns.pairplot(df)




# Plots for Numerical-Categorical Variables

# Boxplot
Description: Displays the distribution of a numerical variable for each category of a categorical variable.

Use Case: Comparing distributions and identifying outliers among categories.
Example:
sns.boxplot(x='categorical_column', y='numerical_column', data=df)

# Violin Plot
Description: Combines a boxplot and KDE plot for each category of a categorical variable.

Use Case: Comparing distributions and densities among categories.
Example:
sns.violinplot(x='categorical_column', y='numerical_column', data=df)

# Barplot
Description: Displays the mean (or other aggregate) of a numerical variable for each category of a categorical variable.

Use Case: Comparing averages across categories.
Example:
sns.barplot(x='categorical_column', y='numerical_column', data=df)

# Stripplot
Description: Plots individual data points of a numerical variable for each category of a categorical variable.

Use Case: Visualizing the distribution of data points within each category.
Example:
sns.stripplot(x='categorical_column', y='numerical_column', data=df)

# Swarmplot
Description: Similar to a stripplot but avoids overlap, providing a clearer view of the distribution.

Use Case: Visualizing the distribution of data points without overlap.
Example:
sns.swarmplot(x='categorical_column', y='numerical_column', data=df)

# Plots for Categorical-Categorical Variables

# Countplot
Description: Displays the count of observations for each category of a categorical variable.

Use Case: Counting occurrences within each category.
Example:
sns.countplot(x='categorical_column', data=df)

# Heatmap
Description: Displays data in a matrix form, using color to represent values.

Use Case: Showing the relationship between two categorical variables.
Example:
crosstab = pd.crosstab(df['categorical_column1'], df['categorical_column2'])
sns.heatmap(crosstab, annot=True, fmt='d')

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Create a crosstab (contingency table)
crosstab = pd.crosstab(titanic['class'], titanic['survived'])

# Plot the heatmap
sns.heatmap(crosstab, annot=True, fmt='d', cmap='coolwarm')
plt.title('Heatmap of Titanic Passenger Class vs. Survival')
plt.xlabel('Survived')
plt.ylabel('Class')
plt.show()
```

# Special Plots

# FacetGrid
Description: Creates a grid of plots based on the values of one or more categorical variables.

Use Case: Analyzing subsets of data across multiple categories.

Example:
g = sns.FacetGrid(df, col='categorical_column1', row='categorical_column2')
g.map(sns.scatterplot, 'numerical_column1', 'numerical_column2')

# PairGrid
Description: Similar to pairplot, but offers more customization options.

Use Case: Detailed exploration of pairwise relationships between numerical variables.

Example:
g = sns.PairGrid(df, hue='categorical_column')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()

# Regression Plots (lmplot)
Description: Fits and visualizes a linear regression model.

Use Case: Analyzing and visualizing the linear relationship between two variables.

Example:
sns.lmplot(x='numerical_column1', y='numerical_column2', hue='categorical_column', data=df)
