```python
import seaborn as sns 
```

```python
data = sns.load_dataset('titanic')
data
```

```python
# Basic strip plot of age by passenger class
sns.stripplot(x="pclass", y="age", hue='survived' ,data=data)

# Show plot
plt.title("Strip Plot of Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

```python
# Strip plot with jitter to avoid overlap
sns.stripplot(x="pclass", y="age", data=data, hue='survived', jitter=True)

# Show plot
plt.title("Strip Plot of Age by Passenger Class with Jitter")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

```python
# Strip plot with hue to differentiate between survival status
sns.stripplot(x="pclass", y="age", hue="survived", data=data, dodge=True)

# Show plot
plt.title("Strip Plot of Age by Passenger Class and Survival Status")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.legend(title="Survived")
plt.show()
```

```python
# Strip plot with customizations (size, palette, marker, edgecolor, alpha)
sns.stripplot(x="pclass", y="age",hue='survived', data=data, jitter=0.3, size=6, palette="Set2", marker="*", edgecolor="gray", alpha=0.9)

# Show plot
plt.title("Customized Strip Plot of Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

# A strip plot is a type of scatter plot that displays the distribution of a numerical variable for different categories by plotting individual data points along an axis. It is useful for visualizing all data points and understanding the spread and distribution within each category.

# Features of Strip Plot
Individual Data Points: Each point represents an individual observation, providing a clear view of the data distribution.

Categorical vs. Numerical: Strip plots are ideal for comparing the distribution of a numerical variable across different categories.

Jitter: Adds random noise to the data points to avoid overlap and improve visibility.

Hue: Allows for additional grouping within categories to show sub-group differences.

# When to Use Strip Plots
Categorical vs. Numerical Data: When you need to visualize the distribution of a numerical variable across different categories.

Detailed Data Inspection: When it's important to see all individual data points rather than summary statistics.

Preventing Overlap: When data points overlap, using jitter can help in distinguishing individual points.

Comparing Sub-Groups: When you want to compare sub-groups within categories using the hue parameter.

# Key Parameters
x and y: Variables to be plotted on the x-axis and y-axis, respectively.
data: DataFrame containing the data to be plotted.

hue: Variable for grouping the data and plotting points with different colors for each group.

jitter: Adds random noise to the data points along the categorical axis to prevent overlap. Can be a boolean or a float specifying the amount of jitter.

dodge: Separates the points along the categorical axis when using hue.

size: Size of the points.

palette: Color palette for different categories or groups.

marker: Shape of the markers.

edgecolor: Color of the point edges.

alpha: Transparency level of the points.
