# Rug Plot
A rug plot is a simple and useful visualization tool that adds small vertical lines (or "rugs") along the x-axis to represent individual data points in a dataset. It is primarily used to show the distribution and density of data points along a single continuous variable.

# Key Characteristics of Rug Plot
Rug Marks: Small vertical or horizontal lines positioned at data points.

Axes: Typically, rug plots are drawn along the x-axis, but they can also be extended to the y-axis.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset("tips")

# Creating a rug plot
sns.rugplot(data["total_bill"])

# Show plot
plt.show()
```

```python
# KDE plot with a rug plot
sns.kdeplot(data["total_bill"], fill=True)
sns.rugplot(data["total_bill"])

plt.show()
```

# Use Cases
Visualizing Distribution: Rug plots are used to provide a visual representation of the distribution of data points along a continuous variable. They complement histograms and KDE plots by showing individual data points.

Identifying Clusters and Outliers: Rug plots help identify clusters of data points and potential outliers. Clusters appear as dense groups of rug marks, while outliers stand alone or in sparse regions.

Complementing Other Plots: Rug plots are often used in conjunction with histograms, KDE plots, and scatter plots to provide a more detailed view of the data distribution.
