```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
flights = sns.load_dataset("flights").pivot(index= "month",columns= "year", values ="passengers")

# Create a basic heatmap
sns.heatmap(flights)

# Show plot
plt.title("Basic Heatmap of Flight Data")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```

```python
# Heatmap with annotations
sns.heatmap(flights, annot=True,fmt="d" )

# Show plot
plt.title("Heatmap of Flight Data with Annotations")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
flights
```

annot=True: Enables annotations, so each cell will display its value.

fmt="d": Formats the annotations as integers.

# Use Cases for Heatmaps
Correlation Matrix: Visualizing the correlation between different variables in a dataset.

Comparing Values Across Categories: Useful for data that can be categorized in a matrix form (e.g., monthly sales data across years).

Clustering: When combined with hierarchical clustering, it shows the similarity between rows and columns.

Data Density: Understanding the density and distribution of data points in a two-dimensional space.

Genomic Data: Widely used in bioinformatics to visualize patterns in genomic data

# When to Use Heatmaps
Matrix-Like Data: When your data can naturally be organized into a matrix format (e.g., time series data, confusion matrices).

Patterns and Trends: When you need to quickly identify patterns, trends, or anomalies in your data.

Comparison Across Multiple Categories: When comparing values across different categories or groups.

Correlation Analysis: When visualizing the correlation between multiple variables.
